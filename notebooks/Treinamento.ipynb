```python
train_df = catalog.load('base_train')
test_df = catalog.load('base_test')
```

```python
# Registrar no MLflow
mlflow.set_experiment("PreparacaoDados")
```

```python
# Tamanhos das bases
train_size = train_df.shape[0]
test_size = test_df.shape[0]

with mlflow.start_run(run_name="PreparacaoDados"):
    mlflow.log_param("test_size", 0.2)
    mlflow.log_metric("train_size", train_size)
    mlflow.log_metric("test_size", test_size)
```

```python
# Obter o ltimo experimento registrado
experiment = mlflow.get_experiment_by_name("PreparacaoDados")
runs = mlflow.search_runs(experiment.experiment_id)

# Mostrar as ltimas execues
print(runs[["run_id", "metrics.train_size", "metrics.test_size", "params.test_size"]])
```
